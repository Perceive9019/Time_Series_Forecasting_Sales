## 4.0 Decomposition

- We can also visualize our data using a method called seasonal decomposition.
- What does it mean to decompose time series data? It means breaking that data into 3 components:
    - **Trend**: the overall direction that the data is traveling (upwards or downwards)
    - **Seasonality**: cyclical patterns in the data
    - **Noise**: the random variation in the data
 
- Is our data conducive to an additive or mutlicative model? To answer this question we read this [article](https://sigmundojr.medium.com/seasonality-in-python-additive-or-multiplicative-model-d4b9cf1f48a7).
    - "In the **additive model**, the behavior is linear where changes over time are consistently made by the same amount, like a linear trend. In this situation, the linear seasonality has the same amplitude and frequency."
    - **Multliplicative model**: the multiplicative model is appropriate when the seasonal variation increases or decreases as the time series data grows. In other words, as the data values increase, the difference between peaks and troughs becomes proportionally wider.
 
- Our data is conducive to a multiplicative model because it has increasing and decreasing amplitudes therefore, the fluctuations seem to be proprotional to the trend and not constant over time. However, we are going to perform a decomposition of the time series data into its trend, seasonality and residuals to comfirm this.

```python
#from statsmodels.tsa.seasonal import seasonal_decompose
#
# Decompose the time series data into trend, seasonality, residual and observed components
decomposition = seasonal_decompose(y, model='multiplicative')
#
# Access and plot individual components
trend_component = decomposition.trend
seasonal_component = decomposition.seasonal
residual_component = decomposition.resid
observed_component = decomposition.observed
#
# Plot each component separately
plt.figure(figsize=(18, 8))

plt.subplot(411) # 4 rows, 1 column, plot 1
plt.plot(observed_component, label='Observed')
plt.legend(loc='best')
plt.title('Observed Component')

plt.subplot(412) # 4 rows, 1 column, plot 2
plt.plot(trend_component, label='Trend', color='blue')
plt.legend(loc='best')
plt.title('Trend Component')

plt.subplot(413) # 4 rows, 1 column, plot 3
plt.plot(seasonal_component, label='Seasonal', color='green')
plt.legend(loc='best')
plt.title('Seasonal Component')

plt.subplot(414) # 4 rows, 1 column, plot 4
plt.plot(residual_component, label='Residual', color='red')
plt.legend(loc='best')
plt.title('Residual Component')

plt.tight_layout()
plt.show()
```

**Note**: from the plot above we can see that our timer series furniture sales data has obvious seasonality.

## 5.0 Testing for Stationarity

Before applying most time series forecasting methods, we need to ensure that our time series data is stationary. Stationarity means that the statistical properties of the series, like mean, variance, and autocorrelation, are constant over time.

### 5.1 ADF Test

The Augmented Dickey-Fuller test is a formal statistical test for stationarity.

- H0: The **null hypothesis** of the test is that the time series can be represented by a unit root, which is **not stationary** (has some time-dependent structure).
- H1: The **alternative hypothesis** is that the time series **is stationary** (does not have a unit root).

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(y)
#
# Print test-statistic
print('ADF Statistic: %f' % result[0])
# 
# Print p-value
print('p-value: %f' % result[1])
#
p_value = result[1]
if p_value <= 0.05:
    print('Data is Stationary')
else:
    print('Data is not stationary')
```

The Augemented Dickey-Fuller (**ADF**) test results indicates the following:
- **ADF Statistic**: a values of -5.321971 is highly negative
- **p-value**: a value of 0.000005 (essentially 0) is far below the common significane level (0.05)

Interpretation:
- Given these results, we reject the null hypothesis of the ADF Test suggesting that the time series furniture data is not stationary.
- This is a good indication we're looking to model our time series with methods like ARIMA or SARIMA. A stationary series often leads to better model performance because the statistical properties (mean, variance, autocorrelation) of the series remain constant over time, which is an assumption in many time series forecasting methods.

### 5.2 Testing for Stationarity with KPSS

Checking for stationarity using both the ADF and KPSS tests provides a more comprehensive view of the data's stationarity.

**KPSS Test**:

- **Null Hypothesis** (H0): The time series is stationary around a deterministic trend.

- **Alternative Hypothesis** (H1): The time series is not stationary (has a unit root).

    - If the p-value is less than a threshold (e.g., 0.05 or 0.01), then we reject the null hypothesis, suggesting the time series is non-stationary.
    - If the p-value is greater than the threshold, we fail to reject the null hypothesis and infer that the time series is stationary.

```python
from statsmodels.tsa.stattools import kpss

# Perform the KPSS test
kpss_result = kpss(y)

# Print the test-statistic
print('KPSS Statistic: %f' % kpss_result[0])

# Print the p-value
print('p-value: %f' % kpss_result[1])

# Print Critical Values
print('Critical Values:', kpss_result[3])
```

**Interpretation of Output Results**:
- **KPSS Statistic (0.193283)**: This is the test statistic calculated by the KPSS test. A lower value suggests a higher likelihood of stationarity.
- **p-value (0.100000)**: This is the closest available p-value, but the actual p-value is higher as shown above in the interpolation warning. In the context of the KPSS test, a high p-value (typically > 0.05) suggests that the null hypothesis of stationarity cannot be rejected.
- **Critical Values**: These are thresholds for the test statistic for different significance levels (10%, 5%, 2.5%, 1%). If our test statistic is higher than these values, it suggests evidence against stationarity at that level of significance. Since our test statistic is lower than all these critical values, it suggests that there is no evidence against the null hypothesis of stationarity at any of these significance levels.

## 6.0 Modeling: Time Series Forecasting with SARIMA

Recall that ARIMA (Autoregressive Integrated Moving Average) models are based around the idea that it's possible to predict the next value in a time series by using information about the most recent data points. It also assumes there will be some randomness in our data that can't ever be predicted.

Since our data exhibits seasonality, we opt for a SARIMA (Seasonal ARIMA) model instead of a standard ARIMA model. SARIMA is an extension of ARIMA that specifically accounts for seasonality in time series data.

### 6.1 Determing the Best Paramteres for an SARIMA Model

- The auto_arima function from the `pmdarima package` in Python is a powerful tool for automatically determining both the non-seasonal and seasonal parameters for a SARIMA model.
- The SARIMA model includes additional seasonal parameters on top of the standard ARIMA parameters:
    - **ARIMA Parameters**:
        - *p* (Autoregressive order): The number of lag observations included in the model (lag order). It signifies the influence of past non-seasonal values on the current value.

        - *d* (Degree of differencing): The number of times that the raw observations are differenced (degree of differencing) to make data stationary (i.e., have constant mean and vairance). It accounts for trends in the data.

        - *q* (Moving average order.): The size of the moving average window (order of moving average). It measures the impact of past forecast errors on the current value.
    - **Seasonal Components of SARIMA**:
        - *P*: Seasonal autoregressive order - this parameter indicates the number of seasonal autoregressive terms in the model. It measures the influence of past seasonal values on the current value.
        - *D*: Seasonal differencing order -  this represents the number of differences required to remove seasonality from the time series. It captures the order of differencing at the seasonal level.
        - *Q*: Seasonal moving average order - this is the number of seasonal moving average terms in the model. It quantifies the impact of past seasonal forecast errors on the current value.
    - **Model Order**:
        - *s*: The number of time steps for a single seasonal period. For example, *s*=12 for monthly data with an annual cycle.

```python
from pmdarima import auto_arima
#
# Recall 'y' with your time series data
#
# Using auto_arima with seasonal parameters
model = auto_arima(y, seasonal=True, m=12, trace=True)
```

```python
print(model.summary())
```

**Results Interpretation**:
- From the results, the model with the lowest AIC value (which generally indicates a better model fit given the trade-off between goodness of fit and model complexity) is the ARIMA(0,0,0)(1,1,0)[12] with an AIC of 488.237. This suggests that this model would be the most suitable for our time series data among the models that were tried.

- We will then fit a SARIMA model with the parameters:

    - Non-seasonal orders:

        - p=0, d=0, q=0,
    - Seasonal orders:

        - P=1, D=1, Q=0
    - Other parameters:

        - m = 12 - seasonal period

### 6.2 Fit SARIMA Model and Training

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Using potentially simplified parameters and enforcing stationarity and invertibility
model = SARIMAX(y, 
                order=(0, 0, 0),  
                seasonal_order=(1, 1, 0, 12),   
                enforce_stationarity=False,
                enforce_invertibility=False)

# Fitting the model with increased iterations
results = model.fit()
print(results.summary().tables[1])
```

```python
# Plot diagnostic charts for our SARIMA model 
results.plot_diagnostics(figsize=(16, 14))
plt.show()
```

The diagnostic plots typically include:

- **Standardized Residuals**: Checking if the residuals are normally distributed and exhibit no patterns (ideally should resemble white noise).
- **Histogram plus Estimated Density**: It shows the kernel density estimate overlaid on a histogram of the standardized residuals. For a good model, this should be approximately normally distributed.
- **Normal Q-Q Plot**: This quantile-quantile plot is used to assess if the distribution of the residuals follows a normal distribution.
- **Correlogram (ACF Plot)**: This plot shows the autocorrelations of the residuals. For a well-fitted model, there should be no significant autocorrelations.

These diagnostic plots are crucial for validating the assumptions of the model and assessing whether the residuals appear as random noise (indicating a good fit) or if they exhibit systematic patterns (indicating a poor fit).

### 6.3 Validating Forecats

To helps us understand the accuracy of our forecats, we compare predicated sales to real observed sales of the time series, and we set forecats to start from 2018-1-01 to the end of the data.

```python
# Get predictions from SARIMA Model
# `get_predictions` method obtains in-sample and out-of-sample forecasts starting from January 1, 2017
pred = results.get_prediction(start=pd.to_datetime('2018-01-01'), dynamic=False)
#
# Calculate the confidence interval for the predictions i.e. the range we expect the true values to lie with a 95% probabilty.
pred_ci = pred.conf_int()
# Plot the observed data from 2015 onwards. This is the historical data used for fitting the SARIMA model
#
ax = y['2015':].plot(label='observed')
# This line plots the predicted mean of the forecasts on the same axes (ax).
# The predictions are labeled as 'One-step ahead Forecast'.
pred.predicted_mean.plot(ax=ax, label='Next Step Forecast', alpha=0.7, figsize=(16, 7))
#
# The `ax.fill_between` part is used to fill the area between the lower and upper bounds of the confidence interval. 
# It shades this area in black (color='k') with 20% opacity (alpha=.2).
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:, 0],
                pred_ci.iloc[:, 1], color='k', alpha=.2)
# Label the x-axis as Date and y-axis as Furniture Sales
ax.set_xlabel('Date')
ax.set_ylabel('Furniture Sales')
plt.legend()
#
plt.show()
```

**Note**: 
- The line plot displays observed furniture sales over time in blue, juxtaposed with forecasted sales in red. The observed data spans from January 2015 to an endpoint in 2018, with the sales figures demonstrating noticeable seasonality, indicated by the regular peaks and troughs throughout the series. The 'Next Step Forecast' starts at the tail end of the observed data, continuing the trend and seasonal pattern into the future.

- The forecast aligns closely with the historical data, suggesting that the model has captured the underlying trend and seasonal fluctuations effectively. This is an indication that the model can be a useful tool for estimating future sales.

- The gray shaded area represents the confidence interval for the forecasts, which provides a range within which future sales are likely to fall, with a typical confidence level of 95%. This interval widens as the forecast extends further from the last observed data point, reflecting increased uncertainty in the longer-term predictions.

### 6.4 Model Evaluation

```python
# Store predicted mean values from the forecasting SARIMA model
y_forecasted = pred.predicted_mean 
#
# stores the actual values from the dataset (`y`) starting from '2018-01-01'
y_truth = y['2018-01-01':]
#
# Compute Mean Square Error
# mse= ((y_forecasted - y_truth) ** 2).mean()
from sklearn.metrics import mean_squared_error
#
mse = mean_squared_error(y_truth, y_forecasted)
#
print('The Mean Squared Error of our Forecasts is {:.2f}'.format(mse))
```

```python
# Calculate RMSE
#  Squreroot of mse i.e., np.sqrt(mse)
#
from math import sqrt
#
rmse = sqrt(mse)
#
# Print the RMSE of the forecasts
print('The Root Mean Squared Error of our forecasts is {:.2f}'.format(rmse))
```

**Results Interpretation**:

- Mean Squared Error:
    - This value (25649.90) represents the average of the squares of the errors, which are the differences between the predicted values and the actual values. 
    - The error in our context is squared to penalize larger errors more severely than smaller ones, which means that a few large errors will result in a much higher MSE compared to many small errors.
    - Therefore, the smaller the MSE, the closer we are to finding the line of best fit.

- Root Mean Squared Error:
    - This is the square root of the MSE and is a more interpretable measure as it is on the same scale as the original data. 
    - An RMSE of 160.16 means that, on average, the model's predictions are off by approximately 160.16 units from the actual values. 
    - Root Mean Square Error (RMSE) tells us that our model was able to forecast the average daily furniture sales in the test set within 160.16 of the actual sales. Our furniture daily sales range from around 200 to over 1600. In my opinion, this is a pretty good model so far.

```python
from prophet import Prophet
```

### 8.2 Prepare Furniture Dataset

```python
# ds = data stamp and y are the values you are forecasting
furniture = furniture.rename(columns={'Order Date': 'ds', 'Sales': 'y'})
furniture.head(2)
```

### 8.3 Create and Fit the Model for Furniture Data

```python
# The `interval_width = 0.95` parameter specifies the uncertainty level (95%).
# The `fit` method is used to fit the model to the furniture dataset.
furniture_model = Prophet(interval_width=0.95)
furniture_model.fit(furniture)
```

### 8.4 Prepare the Office Dataset

```python
# ds = data stamp and y are the values you are forecasting
office = office.rename(columns={'Order Date': 'ds', 'Sales': 'y'})
office.head(2)
```

### 8.5 Create and Fit the Model for Office Data


```python
# The `interval_width = 0.95` parameter specifies the uncertainty level (95%).
# The `fit` method is used to fit the model to the furniture dataset.
office_model = Prophet(interval_width=0.95)
office_model.fit(office)
```

### 8.6 Create Future DataFrames for Prediction (Furniture and Office Data)

```python
# furniture_model is our Prophet model that has already been fitted to the historical furniture data.
# make_future_dataframe(periods=36, freq='MS') is a method used to create a DataFrame
# that extends into the future for a specified number of periods.
# periods=36 indicates that the forecast should extend 36 time periods into the future.
# freq='MS' specifies the frequency of the forecast. 'MS' stands for 'Month Start', 
# meaning each period represents the start of a new month.
furniture_forecast = furniture_model.make_future_dataframe(periods = 36, freq='MS')
```

### 8.7 Making Predictions for Future Furniture Sales and Office Sales

```python
#  The predict method of the fitted furniture_model is applied to the furniture_forecast DataFrame.
# The model uses its learned parameters from the historical data to forecast values for the future dates in furniture_forecast.
# The result is a DataFrame with predicted sales values for the next 36 months.
furniture_forecast = furniture_model.predict(furniture_forecast)
```

### 8.8 Creating Future DataFrames for Prediction (Office Data)

```python
office_forecast = office_model.make_future_dataframe(periods=36, freq='MS')
```

### 8.9 Making Predictions for Future Office Sales

```python
office_forecast = office_model.predict(office_forecast)
```

#### 9.1.1 Create lists of new column names for the furniture_forecast and office_forecast DataFrames.

```python
furniture_names = ['furniture_%s' % column for column in furniture_forecast.columns]
office_names = ['office_%s' % column for column in office_forecast.columns]
```

#### 9.1.2 Creates copies of the furniture_forecast and office_forecast to avoid modifying the original DataFrames

```python
furniture_forecast_copy = furniture_forecast.copy()
office_forecast_copy = office_forecast.copy()
```

#### 9.1.3 The columns of furniture_forecast_copy and office_forecast_copy are renamed using the lists created in 9.1.1 

```python
furniture_forecast_copy.columns = furniture_names
office_forecast_copy.columns = office_names
```

#### 9.1.4 Merge the Two Forecast DataFrames

```python
forecast = pd.merge(furniture_forecast_copy, office_forecast_copy, how = 'inner', left_on='furniture_ds', right_on='office_ds')
forecast.head(2)
```

***Note**: The keys for the merge are specified by left_on = 'furniture_ds' and right_on = 'office_ds'. This means the merge aligns rows where the 'furniture_ds' column in merge_furniture_forecast matches the 'office_ds' column in merge_office_forecast.

#### 9.1.5 Finalize the Merged DataFrame

```python
# The 'furniture_ds' column is renamed to 'Date' for clarity, as it now represents the date for both furniture and office forecasts.
# The 'office_ds' column is dropped (drop('office_ds', axis=1)) since it's redundant after the merge (it's equivalent to 'furniture_ds'/'Date').
forecast = forecast.rename(columns={'furniture_ds':'Date'})
```
